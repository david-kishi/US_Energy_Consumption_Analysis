```python
import requests
import json
import pandas as pd

from config import eia_api_key
```

```python
# All Grades of Gasoline, US Average, Monthly
response = requests.get('http://api.eia.gov/series/?api_key=' + eia_api_key + '&series_id=TOTAL.MGUCUUS.M')
#print(json.dumps(response.json(), indent=4))
response
```

```python
print(json.dumps(response.json()['series'][0]['data'], indent=4))
```

```python
dates = []
all_grade_avg_price = []

for item in response.json()['series'][0]['data']:
    dates.append(item[0])
    all_grade_avg_price.append(item[1])
    
test = pd.DataFrame(all_grade_avg_price, index=dates)
test.columns = ['AllGradeAvgPrice']
test.head()
```
